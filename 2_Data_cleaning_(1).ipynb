```python
import pandas as pd

wikiLink="https://es.wikipedia.org/wiki/%C3%8Dndice_de_pobreza_multidimensional"
tabla1=pd.read_html(wikiLink, flavor='bs4',match="IPM")
len(tabla1)
```

```python
tabla=tabla1[0].copy()
tabla.head()
```

```python

# check headers
tabla.columns
```

```python
nombres_indices = ['País', 'IPM', 'Número de pobres  por IPM (millones)',
                   '% de gente que son pobres IPM', '% Intensidad de Pobreza IPM',
                   '% de gente que son pobres de ingresos ($1.25).']
nuevos_nombres_indices = nombres_indices.copy()
nuevos_nombres_indices[2] = 'N de Pobres'
nuevos_nombres_indices[3] = 'Pobres %'
nuevos_nombres_indices[4] = 'Intensidad Pobreza %'
nuevos_nombres_indices[5] = 'Pobres De Ingreso %'

tabla.columns = nuevos_nombres_indices
print(tabla)

tabla
```

```python
nuevos_nombres_indices

nuevos_nombres_indices = ['País',
 'IPM',
 '# de Pobres',
 '% de Pobres',
 '% Intensidad Pobreza',
 '% Pobres De Ingreso']
nuevos_nombres_indices = [nombre.replace(' ', '') for nombre in nuevos_nombres_indices]
nuevos_nombres_indices
#Lo incluiriamos en la tabla pero podemos ordenarlo de diferente manera
```

```python
def obtener_primeras_letras(texto):
    return ''.join([palabra[0] for palabra in texto.split()])

# Seleccionar las columnas a partir de la tercera en adelante
columnas_a_modificar = tabla.columns[2:]

# Obtener las primeras letras de cada nombre de columna seleccionada y asignarlas como nuevos nombres de columna
nuevos_nombres_columnas = [obtener_primeras_letras(nombre) for nombre in columnas_a_modificar]

# Asignar los nuevos nombres de columna al DataFrame
tabla.columns.values[2:] = nuevos_nombres_columnas

# Imprimir el DataFrame para verificar los cambios
print(tabla)
```

```python
indices= tabla.index
print(indices)
```

```python
import pandas as pd
tabla.IPM.dtype
tabla['IPM'] = tabla.IPM.astype(str)
set(tabla.IPM[~tabla.IPM.str.contains('^\\d+\\.\\d*')])
```

```python
set(tabla.IPM[~tabla.IPM.str.contains('^\\d+\\.\\d*$')])
```

```python
tabla
```

```python


# Aplicamos str.contains a toda la columna 'IPM' del DataFrame
bad_values = tabla.loc[~tabla['IPM'].str.contains(r'^\d+\.*\d*$'), 'IPM'].tolist()

# Elimina los valores duplicados de la lista
bad_values = list(set(bad_values))


bad_values
```

We will need to replace those values with a proper missing value:

```python
import numpy as np

tabla.replace(to_replace=bad_values, value=np.nan,inplace=True)
```

```python
tabla.info
```

Let's keep the complete data:

```python
tabla.dropna(how='any',axis=0,inplace=True, # keep complete
               ignore_index=True) # reset index

tabla
```
